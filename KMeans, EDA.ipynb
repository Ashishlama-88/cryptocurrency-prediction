```python
from sklearn.preprocessing import normalize

df_normal = normalize(df, axis=1)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=10)  # Choose the number of components
df_pca = pca.fit_transform(df_normal)

# X_pca now contains the transformed data with 2 principal components
```

```python
df_pca
```

```python
import numpy as np
from sklearn.cluster import KMeans

# Assuming 'X' is your dataset
k = 4 # Number of clusters
kmeans = KMeans(n_clusters=k, init="k-means++",random_state=10)
clusters = kmeans.fit_predict(df_pca)

# 'clusters' now contains the cluster labels for each data point
```

```python
import matplotlib.pyplot as plt

# Assuming 'X' is your dataset
k = 4  # Number of clusters
kmeans = KMeans(n_clusters=k, init="k-means++", random_state=43)
clusters = kmeans.fit_predict(df_pca)

# Create a scatter plot colored by cluster labels
plt.figure(figsize=(8, 6))
plt.scatter(df_pca[:, 0], df_pca[:, 1], c=clusters, cmap='viridis', edgecolors='k')
plt.colorbar(label='Cluster Number')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Scatter Plot of Data Points Colored by KMeans Clusters')
plt.show()
```
